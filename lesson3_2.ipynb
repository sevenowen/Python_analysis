- 如何依序取出`串列資料`和`映射資料`的元素值
- for_in迴圈

```python
fruits = ['蘋果','香蕉','橘子']
for fruit in fruits:
    print(fruit)
```

```python
for index,fruit in enumerate(fruits):
    print(index, fruit,sep='->')
    print("==========");
```
